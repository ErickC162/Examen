```python
import numpy as np
from scipy.optimize import newton

import matplotlib.pyplot as plt

def polinomio(x):
    return x**5 - 6*x**4 + 2*x**3 + 20*x**2 - 27*x + 10

x = np.linspace(-5,5,1000)

y = polinomio(x)

plt.figure(figsize=(10,6))
plt.ylim(-50,50)
plt.plot(x,y,label=r'$x**5-6*x**4+2*x**3+20*x**2-27*x+10$')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Ecuacion 5° grado")
plt.legend()
plt.show() 
```

```python
def polinomio(x):
    return x**5 - 6*x**4 + 2*x**3 + 20*x**2 - 27*x + 10

def derivada_polinomio(x):
    return 5*x**4 - 24*x**3 + 6*x**2 + 40*x - 27

aproximaciones_iniciales = [-3, -1, 1, 2, 3]

raices = []
for aprox in aproximaciones_iniciales:
    raiz = newton(polinomio, aprox, fprime=derivada_polinomio,maxiter=100)
    raices.append(raiz)

raices_unicas = np.unique(np.round(raices, decimals=10))

print("Raíces encontradas:", raices_unicas)
```
